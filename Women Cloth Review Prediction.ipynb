**Title of Project**: Women's Clothing Review Prediction

**Objective:** The objective of this project is to build a machine learning model that can predict women's clothing reviews based on historical data, helping retailers and manufacturers understand customer sentiments and preferences.

**Data Source:** The data for this project will be sourced from a dataset containing historical women's clothing reviews.



# Import Library

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

# Import Dataset

```python
df = pd.read_csv("https://raw.githubusercontent.com/YBIFoundation/ProjectHub-MachineLearning/main/Women%20Clothing%20E-Commerce%20Review.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

**Missing Values**

```python
df.isna().sum()
```

```python
df[df['Review']==""]=np.NaN
```

```python
df['Review'].fillna("No Review",inplace=True)
```

```python
df.isna().sum()
```

```python
df['Review']
```

**Define Target(y) and feature (X)**

```python
df.columns
```

```python
x = df['Review']
```

```python
y = df['Rating']
```

# Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, stratify=y, random_state=2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Getting Tokens from Feature Text

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(lowercase=True, analyzer='word', ngram_range=(2, 3), stop_words='english', max_features=50000)
```

```python
x_train = cv.fit_transform(x_train)
```

```python
cv.get_feature_names_out()
```

```python
x_train.toarray()
```

```python
x_test = cv.fit_transform(x_test)
```

```python
cv.get_feature_names_out()
```

```python
x_test.toarray()
```

# Modeling

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

# Model Prediction


```python
y_pred = model.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

**Get probability of each predicted class**

```python
model.predict_proba(x_test)
```

#Model Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_test, y_pred))
```

**Recategorizing ratings as Poor (0) and Good (1)**

re-rating 1,2,3 as 0 and 4,5 as 1


```python
df.replace({'Rating': { 1:0, 2:0, 3:0, 4:1, 5:1 }}, inplace=True)
```

```python
y = df['Rating']
```

```python
x = df['Review']
```

#Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, stratify=y, random_state=2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Getting Tokens from Feature Text

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(lowercase=True, analyzer='word', ngram_range=(2, 3), stop_words='english', max_features=50000)
```

```python
x_train = cv.fit_transform(x_train)
```

```python
x_test = cv.fit_transform(x_test)
```

# Model Re-train

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(x_train, y_train)
```

#Model prediction

```python
y_pred = model.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

# Model Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
